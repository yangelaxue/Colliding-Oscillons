```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import os
```

```python
save_dir = "animations"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
key, speed, v = "alpha_beta-0.75_100", "0.005c", "v1"
f_dir = 'alpha_beta-0.75_100_301.15863', 'smash_osc12_osc2_0.005c_v1/larger_box'
```

```python
alpha, beta = key.split('-')[-1].split('_')
alpha, beta = float(alpha), float(beta)
```

```python
image_dir = f"/media/yangelaxue/TRANSFER/Colliding_Oscillons/{f_dir[0]}/{f_dir[1]}"
image_dir
```

```python
# image_fnames_head = "phi slice"
image_fnames_head = "Edens slice"
image_fnames = [f for f in os.listdir(image_dir) if f.startswith(image_fnames_head)]

image_fnames
```

```python
image_times = [float(f[len(image_fnames_head):-4]) for f in image_fnames]
i_sort = np.argsort(image_times)
image_fnames_sorted = [image_fnames[i] for i in i_sort]

image_fnames_sorted
```

```python
images = [plt.imread(os.path.join(image_dir,image_fname)) for image_fname in image_fnames_sorted]
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,6),tight_layout=True)

ax.set_xticks([])
ax.set_yticks([])

ax.set_title(fr"$\alpha,\beta=${alpha},{beta}, v={speed}, {v}")

ims = []
for i in range(len(image_fnames_sorted)):
    im = ax.imshow(images[i], animated=True)
    if i == 0:
        ax.imshow(images[i])  # show an initial one first
    ims.append([im])

# ims = [ax.imshow(plt.imread(os.path.join(image_dir,image_fname)), animated=True) for image_fname in image_fnames_sorted]

ani = animation.ArtistAnimation(fig, ims, interval=300)
ani.save(os.path.join(save_dir,f"{image_fnames_head} {key} {speed} {v} larger box.mp4"),dpi=150)
```
